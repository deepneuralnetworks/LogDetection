```python
import os
import glob
import pandas as pd

from sklearn.metrics import f1_score, accuracy_score
```

```python
# 'DELL LC ', 'xid', 'sxid', 'Infiniband switch', 'HPE IML', 'LSF', 'syslog'
target_type = []
seed_set = [1011, 2022, 3033, 4044, 5055]

# Define F1 method (OPT: [micro, macro])
f1_method = 'macro'

# Define Average Score
average_f1score = 0
average_accscore = 0

for seed in seed_set:
    # Define Best F1 Score
    best_f1 = -1
    
    # List of the result
    lst = glob.glob(f'./result_cls/DAPT_cs_only/{seed}/*')
    
    for i, f in enumerate(lst):
        # Read CSV
        data = pd.read_csv(f)
        
        # Filter Datatype
        if len(target_type) > 0:
            data = data[data.logtype == 'DELL LC ']
        
        # Calculate F1 & ACC
        f1 = f1_score(data.label, data.pred, average=f1_method)
        acc = accuracy_score(data.label, data.pred)
        print(f'[{seed}][Epoch{i}] ACC@ {acc:.3f} | F1@({f1_method}) {f1:.3f}')
        
        # Record Best Result
        if best_f1 < f1:
            best_f1 = f1
            best_seed = seed
            best_epoch = i
            best_acc = acc
    
    # Print Best Score
    print(f'>>> TOP *F1* RESULT >>> [{best_seed}][Epoch{best_epoch}] ACC@ {best_acc:.3f} | F1({f1_method})@ {best_f1:.3f}')
    
    # Record Best Result to average
    average_f1score += best_f1
    average_accscore += best_acc

# Print CV Final Result
print()
print(f'[{len(seed_set)} CV Result] Avg.ACC@ {average_accscore/len(seed_set):.3f} | Avg.F1({f1_method})@ {average_f1score/len(seed_set):.3f}' )
```
